Exemple programa que demani una nota (0-10) i digui si és:

“suspès” (0-4)

“aprovat” (5-6)

“notable” (7-8)

“excel·lent” (9-10)

```python
nota = int(input("Introdueix la nota (0-10): "))
if nota <= 4:
    print("Suspès")
elif nota <= 6:
    print("Aprovat")
elif nota <= 8:
    print("Notable")
else:
    print("Excel·lent")
```

Bucles

for: recórrer rangs i llistes

while: repetir fins a complir una condició

```python
#Imprimir numeros del l'1 al 10
for i in range(1, 11):
    print(i)
```

```python
#Compta enrere de 10 fins a 1
i = 10
while i >= 1:
    print(i)
    i -= 1
```

```python
# Demana dos números a l'usuari i mostra la seva suma.
num1 = int(input("Introdueix el primer número: "))
num2 = int(input("Introdueix el segon número: "))
suma = num1 + num2
print("La suma és:", suma)
```

```python
# Demana el preu d'un producte i aplica un descompte del 10% si el preu és superior a 50.
preu = float(input("Introdueix el preu del producte: "))
if preu > 50:
    preu_final = preu * 0.9
else:
    preu_final = preu
print("El preu final és:", preu_final)
```

```python
# Calcula la suma dels números de l'1 al 5 utilitzant un bucle for.
suma = 0
for i in range(1, 6):
    suma += i
print("La suma és:", suma)
```

```python
# Demana una paraula a l'usuari i imprimeix cada lletra per separat.
paraula = input("Introdueix una paraula: ")
for lletra in paraula:
    print(lletra)
```
